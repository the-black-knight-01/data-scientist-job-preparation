Minimize $ f(x, y) = (x-1)^2 + (y-2.5)^2 $ subjects to

$$ x-2y+2 \geq 0 $$
$$ -x-2y+6 \geq 0 $$
$$ -x + 2y + 2 \geq 0 $$
$$ x \geq 0 $$
$$ y \geq 0 $$



* 2D function takes in vector $x$
* Constraints must be specified as $g_i(x) \geq 0$
* Bounds specified as rectangular 

```python
f = lambda x: (x[0] - 1)**2 + (x[1] - 2.5)**2
```

```python
cons = ({"type": "ineq", "fun": lambda x: x[0] - 2*x[1] + 2},
        {"type": "ineq", "fun": lambda x: -x[0] - 2*x[1] + 6},
        {"type": "ineq", "fun": lambda x: -x[0] + 2*x[1] + 2}
       )
```

```python
bnds = (
        (0, None),
        (0, None)
       )
```

```python
result_2D = minimize(f, (2,3), bounds=bnds, constraints=cons)
```

```python
result_2D
```

```python
result_2D.fun
```

```python
result_2D.x
```

The optimization problem is solved using the SLSQP method as:

```python
result_2D = minimize(f, (2,3), bounds=bnds, constraints=cons, method = "SLSQP")
```

```python
result_2D
```

```python
result_2D.x
```

# Interpolation
- In the mathematical field of numerical analysis, interpolation is a type of estimation, a method of constructing new data points based on the range of a discrete set of known data points.
- https://docs.scipy.org/doc/scipy/reference/interpolate.html

Suppose you have the following data (I generate it here, but perhaps you collect $x$ and $y$ by means of some experiment)

```python
x = np.linspace(0, 10, 10) # this is linear space not line space
```

```python
y = x**2 * np.sin(x)
```

```python
plt.scatter(x, y)
```

```python
from scipy.interpolate import interp1d
```

```python
f = interp1d(x, y, kind="linear")
```

```python
x2 = np.linspace(0, 10, 100)
```

```python
y2 = f(x2)
```

```python
plt.plot(x2, y2)
plt.scatter(x2, y2)
```

```python
f = interp1d(x, y, kind="cubic")
x2 = np.linspace(0, 10, 100)
y2 = f(x2)
plt.plot(x2, y2)
plt.scatter(x2, y2)
```

The equation for spring motion is $y(t) = A\cos(\omega t+\phi)$. Want to find the natural frequency of oscillation $\omega$ for the spring. You collect the data

```python
t_data = np.array([ 0.   ,  0.34482759,  0.68965517,  1.03448276,  1.37931034,
        1.72413793,  2.06896552,  2.4137931 ,  2.75862069,  3.10344828,
        3.44827586,  3.79310345,  4.13793103,  4.48275862,  4.82758621,
        5.17241379,  5.51724138,  5.86206897,  6.20689655,  6.55172414,
        6.89655172,  7.24137931,  7.5862069 ,  7.93103448,  8.27586207,
        8.62068966,  8.96551724,  9.31034483,  9.65517241, 10.        ])
y_data = np.array([ 4.3303953 ,  1.61137995, -2.15418696, -3.90137249, -1.67259042,
        2.16884383,  3.86635998,  1.85194506, -1.8489224 , -3.96560495,
       -2.13385255,  1.59425817,  4.06145238,  1.89300594, -1.76870297,
       -4.26791226, -2.46874133,  1.37019912,  4.24945607,  2.27038039,
       -1.50299303, -3.46774049, -2.50845488,  1.20022052,  3.81633703,
        2.91511556, -1.24569189, -3.72716214, -2.54549857,  0.87262548])
```

$\omega = 2 \pi f$, $f=1/T$, and $T \approx 2$ seconds. Thus good initial guess is
* $\omega = 2 \pi (1/2) = \pi$
* $A=4$
* $\phi = 0$

```python
def f(t, A, w, phi):
    return A*np.cos(w*t + phi)
```

```python
popt, pcov = curve_fit(f, t_data, y_data, p0=(4, np.pi, 0))
```

```python
popt
```

```python
pcov
```

The estimated error on the parameters

```python
error = np.sqrt(np.diag(pcov))
error
```

The parameters

```python
A, w, phi = popt
print(A, w, phi)
```

```python
f()
```

```python
t = np.linspace(0, 10, 100)
y = f(t, A, w, phi)
```

```python
plt.plot(t, y)
```

```python
plt.plot(t, y) # out curve
plt.scatter(t_data, y_data) # data
```

```python
popt, pcov = curve_fit(f, t_data, y_data, p0=(4, 5, 9))
A, w, phi = popt
error = np.sqrt(np.diag(pcov))
error
```

```python
y = f(t, A, w, phi)
plt.plot(t, y) # out curve
plt.scatter(t_data, y_data) # data
```

## Differentiation

$ f(x) = x^2\sin(2x)e^{-x}$

```python
from scipy.misc import derivative
```

```python
x = np.linspace(0,1, 100)
```

```python
def f(x):
    return x**2*np.sin(2*x)*np.exp(-x)
```

```python
#derivative(f, x, dx=1e-6)
```

```python
plt.plot(x, f(x), color='blue')
plt.plot(x, derivative(f, x, dx=1e-6), color='green')
plt.plot(x, derivative(f, x, dx=1e-6, n=2), color='red')
plt.grid()
```

## Integration
- https://docs.scipy.org/doc/scipy/reference/integrate.html

Single integrals

$$ \int_0^{1} x^2\sin(2x)e^{-x}dx $$

```python
from scipy.integrate import quad
```

```python
integrand = lambda x: x**2*np.sin(2*x)*np.exp(-x)
```

```python
integral , integral_error = quad(integrand,  0, 1)
```

```python
integral
```

```python
integral_error
```

Double integrals
$$ \int_0^{1} \int_{-x}^{x^2} \sin(x+y^2)dydx $$

```python
from scipy.integrate import dblquad
```

```python
integrand = lambda x, y: np.sin(x+y**2)
lwr_b = lambda x: -x
upr_b = lambda x: x**2
```

```python
integral, integral_error = dblquad(integrand, 0, 1, lwr_b, upr_b)
```

```python
integral
```

```python
integral_error
```
